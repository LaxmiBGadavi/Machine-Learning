# model evaluation

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
import numpy as np

# Predictions
y_train_pred = model.predict(xtrain)
y_test_pred = model.predict(xtest)

# R² Score
train_score = r2_score(ytrain, y_train_pred)
test_score = r2_score(ytest, y_test_pred)

print("Train R² Score:", train_score)
print("Test R² Score:", test_score)

# Error Metrics
mae = mean_absolute_error(ytest, y_test_pred)
mse = mean_squared_error(ytest, y_test_pred)
rmse = np.sqrt(mse)

print("MAE :", mae)
print("MSE :", mse)
print("RMSE:", rmse)

"""Fit Analysis
Training Accuracy: 96.45%
Testing Accuracy: 90.24%
The difference between training and testing scores is only about 6.2%, which is small.
This indicates that the model performs well on both training and unseen data.
Therefore, the model has low bias and low variance, indicating good generalization.
Conclusion

The Linear Regression model achieved a Train R² Score of 96.45% and a Test R² Score of 90.24%, showing that it performs 
well on both the training and testing datasets. The small difference between the two scores indicates that the model is not 
underfitting or overfitting. The MAE (6286.45) and RMSE (7059.04) are reasonably low, meaning the prediction errors are relatively small.
Hence, the model is considered a Good Fit, with low bias, low variance, and good generalization on unseen data."""
```
